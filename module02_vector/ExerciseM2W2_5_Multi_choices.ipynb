```python
import numpy as np

def compute_vector_length(vector):
    len_of_vector = np.linalg.norm(vector)
    return len_of_vector

vector = np.array([-2, 4, 9, 21])
result = compute_vector_length([vector])
print(round(result, 2))
```

```python
import numpy as np

def compute_dot_product(vector1, vector2):
    result = np.dot(vector1,vector2)
    print(round(result,2))

vector1 = np.array([0, 1, -1, 2])
vector2 = np.array([2, 5, 1, 0])
compute_dot_product(vector1, vector2)
```

```python
def inverse_matrix(matrix):
    det = np.linalg.det(matrix)
    print(det)
    if det ==0:
        print("Matrix is not invertible")
        return None

    matrix2 = np.zeros((2,2))
    matrix2[0,0] = matrix[1,1]
    matrix2[0,1] = -matrix[0,1]
    matrix2[1,0] = -matrix[1,0]
    matrix2[1,1] = matrix[0,0]

    inverse_matrix = matrix2/det
    return inverse_matrix

matrix = np.array([[-2,6],[8,-4]])
inverse_matrix(matrix)
```

```python
def compute_eigenvalues_eigenvectors(matrix):
    eigenvalues, eigenvectors = np.linalg.eig(matrix)
    return eigenvalues, eigenvectors

matrix = np.array([[0.9, 0.2], [0.1, 0.8]])
eigenvalues, eigenvectors = compute_eigenvalues_eigenvectors(matrix)
print(eigenvectors)
```

```python
def compute_cosine(v1, v2):
    if np.linalg.norm(v1) == 0 or np.linalg.norm(v2) == 0:
        result =  "None"
    else:
        result = round(np.dot(v1,v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)),3)
    return result

vector1 = np.array([1, 2, 3, 4])
vector2 = np.array([1, 0, 3, 0])
print(compute_cosine(vector1, vector2))
```
