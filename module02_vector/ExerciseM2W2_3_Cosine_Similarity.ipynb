* Vector x, y:
$x =
\begin{bmatrix}   
1\\
2\\
3\\
4\\
\end
{bmatrix},
y =
\begin{bmatrix}   
1\\
0\\
3\\
0\\
\end
{bmatrix}$

* Cosine Similarity:
$cs(x,y) = {{x.y} \over {||x||.||y||}} = {\sum_1^n{x_i.y_i} \over \sqrt{\sum_1^nx_i^2}\sqrt{\sum_1^ny_i^2}} = {{1*1  + 2*0 + 3*3 + 4*0} \over \sqrt{1^2+2^2+3^2+4^2}\sqrt{1^2+0^2+3^2+0^2}} = 0.58
$

```python
import numpy as np

def compute_cosine(v1, v2):
    if np.linalg.norm(v1) == 0 or np.linalg.norm(v2) == 0:
        result =  "None"
    else:
        result = round(np.dot(v1,v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)),2)
    return result

vector1 = np.array([1, 2, 3])
vector2 = np.array([4,5,6])
print(compute_cosine(vector1, vector2))
```
