$A =

Identity \space matrix \space I =
\begin{bmatrix}

* Eigenvalue (λ): $det(A − λI) = det(\begin{bmatrix}

\end{bmatrix} − λ\begin{bmatrix}   
1 & 0\\
0 & 1\\
\end
{bmatrix}) = 0 \implies λ = 1 \space or \space λ = 7$

* Eigenvector (v): Av = λv ⇐⇒ (A − λI)v = 0

$\begin{bmatrix}

```python
import numpy as np

def compute_eigenvalues_eigenvectors(matrix):
    eigenvalues, eigenvectors = np.linalg.eig(matrix)

    print("Eigenvalues:")
    for eigenvalue in eigenvalues:
        print(eigenvalue)

    print("\nEigenvectors:")
    for eigenvector in eigenvectors:
        print(eigenvector)

A = np.array([[4,1], [2,3]])
compute_eigenvalues_eigenvectors(A)
```
